```python
import h5py
from pathlib import Path
```

```python
catalog_path=Path("/orcd/data/mvogelsb/005/Lumina/Lumina_above_z_4p75/group_files/fof_subhalo_tab_036.hdf5")
```

```python
print(catalog_path.is_file())
```

```python
with h5py.File(catalog_path,"r") as catalog:
    print("groups:", list(catalog.keys()))
    print("header:", dict(catalog["Header"].attrs))
    
    for dataset_path in [
        "Group/GroupPos",
        "Group/Group_M_Crit200",
        "Group/GroupLenType",
    ]:
          dataset=catalog[dataset_path]
          
          print(f"\n{dataset_path}")
          print("shape:", dataset.shape)
          print("dtype:", dataset.dtype)
          print("attribute:", dict(dataset.attrs))
```

```python
with h5py.File(catalog_path,"r") as catalog:
    for name, value in sorted(catalog["Parameters"].attrs.items()):
        print(f"{name}:{value}")
    
```

```python
with h5py.File(catalog_path,"r") as catalog:
    print("Positions:")
    print(catalog["Group/GroupPos"][:5])
    
    print("\nM200c")
    print(catalog["Group/Group_M_Crit200"][:5])
                  
    print("\nParticle count by type:")
    print(catalog["Group/GroupLenType"][:5])
```
